```python
import os
import pandas as pd
import numpy as np
```

```python
path = os.getcwd()
```

```python
df1 = pd.read_excel(path + '/data/dataset-easy-twitter-search-scraper-1.xlsx')
df2 = pd.read_excel(path + '/data/dataset-easy-twitter-search-scraper-2.xlsx')
df3 = pd.read_excel(path + '/data/dataset-easy-twitter-search-scraper-3.xlsx')
```

```python
all(df1.columns == df3.columns) & all(df1.columns == df2.columns)
```

```python
# merging all three dataframes
dfs = pd.concat([df1, df2, df3], axis = 0)
dfs.reset_index(drop = True, inplace = True)
dfs.head(5)
```

```python
# extracting needed columns
dfs = dfs.loc[:, ['id', 'likes', 'replies', 'retweets', 'text', 'timestamp', 'url']]
dfs.drop_duplicates(subset = 'id', keep = 'first', inplace = True) # dropped about 2k duplicated tweets
dfs.set_index('id', inplace = True)
dfs.head(5)
```

```python
dfs.shape
```

```python
# reading two more datasets formatted in a different way, going to clean and adjust format in the following
dflash1 = pd.read_csv('/home/jovyan/india-research/data/dataset_tweet-flash_1.csv')
dflash2 = pd.read_excel('/home/jovyan/india-research/data/dataset_tweet-flash_2.xlsx')
```

```python
# clean the two datasets we just read
dflash1 = dflash1.loc[:, ['likes', 'replies', 'retweets', 'text', 'timestamp', 'url']]
dflash1['id'] = dflash1['url'].str[-19:].astype(int)
dflash1.set_index('id', inplace = True)

dflash2 = dflash2.loc[:, ['likes', 'replies', 'retweets', 'text', 'timestamp', 'url']]
dflash2['id'] = dflash2['url'].str[-19:].astype(int)
dflash2.set_index('id', inplace = True)
```

```python
# reading and cleaning the last dataset that has another format as well
dfurl = pd.read_csv('/home/jovyan/india-research/data/dataset_twitter-url-scraper-1.csv')
dfurl = dfurl.loc[:, ['favorite_count', 'reply_count', 'retweet_count', 'full_text', 'created_at', 'url']]
dfurl = dfurl.rename(columns = {'favorite_count': 'likes', 'reply_count': 'replies', 'retweet_count': 'retweets',
                               'full_text': 'text', 'created_at': 'timestamp'})
dfurl['id'] = dfurl['url'].str[-19:]
dfurl['id'] = dfurl['id'].str.lstrip('/')
dfurl['id'].astype(int)
dfurl.set_index('id', inplace = True)
```

```python
# let's now merge the two dflashes, dfurl, and the dfs and store it as df variable
df = pd.concat([dfs, dfurl, dflash1, dflash2], axis = 0)

# let's drop duplicates
df = df[~df.index.duplicated(keep = 'first')]
df.head(5)
```

```python
# adjust timestamps across the dataframe
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
df.shape
```

```python
df.to_csv('final_df.csv', index = False)
```

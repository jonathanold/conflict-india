#### In this notebook, I will read, clean, and merge 16 different csv and excel files containing datasets of tweets scraped from the @BloodDonorsIn. I will save the resulting dataset into a csv file called final_df.csv.

```python
import os
import pandas as pd
import numpy as np
```

```python
path1 = os.getcwd()
print(path1)
path = os.path.join(os.path.dirname(os.path.dirname(os.path.dirname(os.path.dirname(path1)))), "Data/Twitter/")
print(path)
```

```python
file_names = ["dataset_easy-twitter-search-scraper_2023-09-18_07-11-32-089.xlsx",
    "dataset_easy-twitter-search-scraper_2023-09-18_08-26-29-038.xlsx",
    "dataset_easy-twitter-search-scraper_2023-09-18_18-52-39-192.xlsx"]

df1 = pd.read_excel(path + file_names[0])
df2 = pd.read_excel(path + file_names[1])
df3 = pd.read_excel(path + file_names[2])
```

```python
all(df1.columns == df3.columns) & all(df1.columns == df2.columns)
```

```python
# merging all three "easy-twitter-search-scraper" dataframes
dfset1 = pd.concat([df1, df2, df3], axis = 0)
dfset1.reset_index(drop = True, inplace = True)
dfset1.head(5)
```

```python
# extracting needed columns
dfset1 = dfset1.loc[:, ['id', 'likes', 'replies', 'retweets', 'text', 'timestamp', 'url']]
dfset1.drop_duplicates(subset = 'id', keep = 'first', inplace = True) # dropped about 2k duplicated tweets
dfset1.set_index('id', inplace = True)
dfset1.head(5)
```

```python
dfset1.shape
```

```python
# reading 11 xlsx and 8 csv tweet-flash-scraped datasets, formatted in a different way, going to clean and adjust format in the following:
file_names_xlsx = [
    "dataset_tweet-flash_2023-09-22_10-51-50-140.xlsx",
    "dataset_tweet-flash_2023-10-07_02-21-59-781.xlsx",
    "dataset_tweet-flash_2023-10-07_04-20-17-613.xlsx",
    "dataset_tweet-flash_2023-10-07_21-13-07-024.xlsx",
    "dataset_tweet-flash_2023-10-08_08-53-21-926.xlsx",
    "dataset_tweet-flash_2023-10-08_09-45-08-890.xlsx",
    "dataset_tweet-flash_2023-10-08_09-45-09-775.xlsx",
    "dataset_tweet-flash_2023-10-08_21-27-51-899.xlsx",
    "dataset_tweet-flash_2023-10-08_22-01-26-250.xlsx",
    "dataset_tweet-flash_2023-10-09_00-41-48-717.xlsx",
    "dataset_tweet-flash_2023-10-09_05-44-22-845.xlsx"]

file_names_csv = [
    "dataset_tweet-flash_2023-09-16_02-03-32-004.csv",
    "dataset_tweet-flash_2023-10-20_18-43-29-410.csv",
    "dataset_tweet-flash_2023-10-20_19-27-01-605.csv",
    "dataset_tweet-flash_2023-10-20_19-40-11-196.csv",
    "dataset_tweet-flash_2023-10-20_19-42-53-952.csv",
    "dataset_tweet-flash_2023-10-20_19-43-29-444.csv",
    "dataset_tweet-flash_2023-10-20_19-46-01-086.csv",
    "dataset_tweet-flash_2023-10-20_19-53-42-590.csv"]

# creating a second dfset and putting xlsx and csv files with two for loops in order
dfset2 = {}
for i in range(len(file_names_xlsx)):
    dflashi = pd.read_excel(path + file_names_xlsx[i])
    dfset2[f'dflash{i}'] = dflashi
            
for i in range(len(file_names_csv)):
    dflashi = pd.read_csv(path + file_names_csv[i])
    dfset2[f'dflash{i+11}'] = dflashi
    
print("Done")
```

```python
# make the dfset2 files to be same format as dfset1
for i in range(19):
    dfset2[f'dflash{i}'] = dfset2[f'dflash{i}'].loc[:, ['likes', 'replies', 'retweets', 'text', 'timestamp', 'url']]
    dfset2[f'dflash{i}']['id'] = dfset2[f'dflash{i}']['url'].str[-19:]
    dfset2[f'dflash{i}']['id'] = dfset2[f'dflash{i}']['id'].str.lstrip('/').astype(int)
    dfset2[f'dflash{i}'].set_index('id', inplace = True)
```

```python
# reading and cleaning the last dataset that has another format
dfurl = pd.read_csv(path + 'dataset_twitter-url-scraper_2023-09-16_00-19-46-917.csv')
dfurl = dfurl.loc[:, ['favorite_count', 'reply_count', 'retweet_count', 'full_text', 'created_at', 'url']]
dfurl = dfurl.rename(columns = {'favorite_count': 'likes', 'reply_count': 'replies', 'retweet_count': 'retweets',
                               'full_text': 'text', 'created_at': 'timestamp'})
dfurl['id'] = dfurl['url'].str[-19:]
dfurl['id'] = dfurl['id'].str.lstrip('/')
dfurl['id'].astype(int)
dfurl.set_index('id', inplace = True)
```

```python
# let's now merge these the dflash-formatted files that are in a dictionary into a new dfset3
dfset3 = pd.concat(dfset2.values(), axis = 0)
dfset3

# merge that with dfurl, and the dfs and store it as df variable
df = pd.concat([dfset1, dfset3, dfurl], axis = 0)
df

# let's drop duplicates
df = df[~df.index.duplicated(keep = 'first')]
df.head()
```

```python
# let's transform timestamp values to datetime values and extract dates for future visualizations 
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['date'] = df['timestamp'].dt.date
df
```
